17. Build chains of trips. Two trips are consecutive in a chain if (a) they have the same VendorID, (b) the pickup location of the second trip is also the dropoff location of the first trip, (c) the pickup time of the second trip is after the dropoff time of the first trip, and (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip.

Hint: Add a column chain to the dataset. A chain can have more than two trips.



```python
data17 = data.copy()
data17 = data[['VendorID', 'tpep_pickup_datetime', 'tpep_dropoff_datetime', 'PULocationID', 'DOLocationID']]


# Ordiniamo i dati per t_pep_pickup_datetime
data17_sorted = data17.sort_values('tpep_pickup_datetime').reset_index(drop=True).copy()
data17_sorted['id_'] = data17_sorted.index  # ID per ciascun viaggio 
data17_sorted['pickup_key'] = data17_sorted['VendorID'].astype(str) + '_' + data17_sorted['PULocationID'].astype(str)
```

```python
# Dataframe 'drop' per effettuare successivamente il merging dei dati 
data17_drop = data17_sorted[['id_', 'tpep_dropoff_datetime', 'VendorID', 'DOLocationID']].copy()
data17_drop['pickup_key'] = data17_drop['VendorID'].astype(str) + '_' + data17_drop['DOLocationID'].astype(str)
data17_drop = data17_drop.sort_values('tpep_dropoff_datetime')
```

```python
# Merge_asof per cercare i viaggi compatibili rispetto alle località PUL e DOL
data17_merged = pd.merge_asof(
    data17_sorted,
    data17_drop[['id_', 'tpep_dropoff_datetime', 'pickup_key']].rename(columns={'id_': 'id_prev'}),
    left_on='tpep_pickup_datetime',
    right_on='tpep_dropoff_datetime',
    by='pickup_key',
    direction='backward',
)
data17_merged
```

```python
# Rinomina la colonna  tpep_dropoff_datetime_y con  prev_dropoff_time
data17_merged.rename(columns={'tpep_dropoff_datetime_y': 'prev_dropoff_time'}, inplace=True)

# Calcola la differenza tra  il corrente viaggio da pickup and e il precedente viaggio che hanno lo stesso dropoff
data17_merged['time_diff'] = data17_merged['tpep_pickup_datetime'] - data17_merged['prev_dropoff_time']
data17_merged
```

```python
# Mantiene solo le relazioni valide 
data17_merged.loc[
    (data17_merged['prev_dropoff_time'].isna()) | 
    (data17_merged['time_diff'] <= pd.Timedelta(0)) | 
    (data17_merged['time_diff'] > pd.Timedelta(minutes=2)), 'id_prev'] = -1

data17_merged
```

```python
# Usa un dizionario per tenere traccia dell'inzio di ciascuna catena 
chains = {}
# Crea un dizionario che mappi ogni id al suo precedente viaggio
prev_dict = dict(zip(data17_merged['id_'], data17_merged['id_prev']))
```

```python
# Inizia a costruire le chain 
for id_ in data17_merged['id_']:
    current_id = id_    # tiene traccia del viaggio in corso
    chain_members = []  # lista per memorizzare tutti gli id nella stessa chain 
    
    # Se la lista raggiunge  -1 significa che non ci sono viaggi precedenti e consecutivi 
    while current_id in prev_dict and prev_dict[current_id] != -1:
        chain_members.append(current_id)
        current_id = prev_dict[current_id]

    
    # Aggiunge  il primo elemento della catena 
    chain_members.append(current_id)

    # Crea un unico chain ID che assegna a tutti i membri della catena 
    chain_id = len(chains) + 1
    for member in chain_members:
        chains[member] = chain_id

# Questo comando assegna alla colonna 'chain' i valori: sostituisce ogni 'id_' con il corrispondente valore in chains 
data17_merged['chain'] = data17_merged['id_'].map(chains) 
```

```python
# Conta le occorrenze di ciascuna catena  e le memorizza in una nuova colonna 'chain_count'
data17_merged['chain_count'] = data17_merged.groupby('chain')['chain'].transform('count') 
data17_merged
```

```python
# Mantiene solo le catene con almeno 2 viaggi ed elimina i NaN nella catena 
data17_final = data17_merged[data17_merged['chain_count'] >= 2]

unique_chains = sorted(data17_final['chain'].unique())  
chain_mapping = {old_id: new_id for new_id, old_id in enumerate(unique_chains, start=1)} 
data17_final['chain'] = data17_final['chain'].map(chain_mapping) 

# Ordina  per la colonna 'chain' e  drop-off datetime 
data17_final = data17_final.sort_values(by=['chain', 'tpep_dropoff_datetime_x']).reset_index(drop=True)

# Elimina le colonne id_ , pickup_key, id_prev, prev_dropoff_time
data17_final = data17_final.drop(columns=['id_', 'pickup_key', 'id_prev', 'prev_dropoff_time'])

# Rinomina la colonna drop-off datetime 
data17_final.rename(columns={'tpep_dropoff_datetime_x': 'tpep_dropoff_datetime'}, inplace=True)

# Dataframe finale 
data17_final
```
